```python
import pandas as pd
```

```python
df_movies = pd.read_csv('archive/movies.csv')
```

```python
df_movies = df_movies[['title','year','genre','actors','description']]
```

```python
df_movies['year'] = df_movies['year'].replace('TV Movie 2019', 2019)
```

```python
df_movies['year'] = pd.to_numeric(df_movies['year'])
```

```python
df_movies["year"].unique()
```

```python
df_movies = df_movies[df_movies['year'] >= 1980]
```

```python
df_movies["year"].unique()
```

```python
df_movies = df_movies.sort_values('year')
```

```python
df_movies.tail()
```

```python
actors = []
for index, row in df_movies.iterrows():
    if (type(row['actors']) == str):
        actors_in_movie = row['actors']
        vector = actors_in_movie.split(', ')
        for act in vector:
            actors.append(act)
        
actors_set = set(actors)
```

```python
graph = {}
graph_keys = graph.keys()
graph_values = graph.values()
count = 0
for index, row in df_movies.tail().iterrows():
    if (type(row['actors']) == str):
        actors_in_movie = row['actors']
        print('actors_in_movie:', actors_in_movie)
        title = row['title']
        print('title:', title)
        vector = actors_in_movie.split(', ')
        print('vector:', vector)
        
        for actor in vector:
            print('actor:', actor)
            
            if not actor in graph:
                graph[actor] = []
                count = count + 1
                print('count:', count)
                print('graph[actor]:', graph)
                
            other_actors = vector[:]
            other_actors.remove(actor)
            print('other_actors:', other_actors)
            
            for other_actor in other_actors:
                print('other_actor:', other_actor)

                if any(x == other_actor for x in graph[actor]):
                       for item in graph[actor]:
                            print('ITEM:', item)
                            if item['name'] == other_actor:
                                item['shared_movies'] = item['shared_movies'].append(title)
                else:
                    obj = {'name': other_actor,
                           'shared_movies': [title]}
                    graph[actor] = obj
        
# graph = {
#     'Kamen Gabriel': [{'name':'Christina Rose', 'shared_movies': ['harryPotter', 'fepas e fabi uma hist de amor',]}, {}, {}],
#     'actor': [{'name':'actor', 'shared_movies': [,]}],
# }

            
# graph = {
#     'Dave Mantel': [['Julia Batelaan','Kill Mode'], ['Yasmin Blake','Kill Mode']],
#     'v2': ['v3', 'v5', 'v6'],
#     'v3': ['v4', 'v5'],
#     'v4': ['v5'],
#     'v5': ['v7'],
#     'v6': ['v7'],
#     'v7': []
# }
```
